```python
# Take input from user
num1=float(input("Enter first num:"))
num2=float(input("Enter second num:"))
#Perform add operation
sum=num1+num2
#Perform subtract operation
diff=num1-num2
#Perform product operation
mul=num1*num2
#Perform power operation
power=num1**num2
#Print the results
print("sum:",sum)
print("diff:",diff)
print("mul:",mul)
print("power:",power)
```

```python
# Taking input from user
string = list("Python")
# Delete 'h' and 'o' characters
string = [x for x in string if x not in ('h','o')]
# Reverse the resultant string
string = string[::-1]
# Print the reversed string
print("".join(string))
```
